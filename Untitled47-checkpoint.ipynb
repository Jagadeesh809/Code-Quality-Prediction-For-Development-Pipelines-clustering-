# CODE QUALITY PREDICTION FOR DEVELOPMENT PIPELINES (clustering)

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

```

```python
# Sample dataset: Code metrics
data = {
    "Lines_of_Code": [100, 200, 150, 300, 400, 500, 600],
    "Cyclomatic_Complexity": [5, 10, 7, 20, 25, 30, 35],
    "Maintainability_Index": [75, 60, 70, 50, 40, 30, 20],
}

# Create DataFrame
df = pd.DataFrame(data)

```

```python
# Preprocess data: Standardize features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
```

```python

# Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(scaled_data)
```

```python

# Add cluster labels to the DataFrame
df['Quality_Cluster'] = clusters

# Display results
print(df)

```

```python
# Visualize clusters
plt.figure(figsize=(8, 6))
plt.scatter(df['Lines_of_Code'], df['Cyclomatic_Complexity'], c=df['Quality_Cluster'], cmap='viridis')
plt.title('Code Quality Clustering')
plt.xlabel('Lines of Code')
plt.ylabel('Cyclomatic Complexity')
plt.colorbar(label='Quality Cluster')
plt.show()
```
